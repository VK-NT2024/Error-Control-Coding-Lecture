# 2. Error Probabilities and Mutual Information for Digital Linear Modulation Schemes

## 2.1 System Model

From the first exercise, we know the basic model of a point-to-point communication link for linear modulation schemes and an AWGN channel. The received symbol at an arbitrary time instance becomes

\begin{equation}
    y = x + w 
\end{equation}

The optimal Maximum A Posteriori (MAP) detector 

\begin{equation}
    \hat{x}^{\mathsf{MAP}}
    = \argmax_{\tilde{x} \in \XX} \Pr\{ \tilde{x} \mid y \}
    = \argmax_{\tilde{x} \in \XX} \Big[ p(y \mid \tilde{x}) \cdot \Pr\{ \tilde{x}\} \Big] 
\end{equation}

maximizes the a posteori probability $\Pr\{ \tilde{x} \mid y \}$. If all $M$ data symbols are equally likely, the a priori probability $\Pr \{ x \} = \frac{1}{M}$ becomes a contant term and can be skipped. This leads to the maximum likelihood detector

\begin{equation}
    \hat{x}^{\mathsf{ML}} = \argmax_{\tilde{x} \in \XX} p(y \mid \tilde{x}) \; .
\end{equation}

The above expressions for MAP and ML detection can be further rewritten for the assumption of an AWGN channel (additive white Gaussian noise). Without loss of generality, the natural logarithm can be applied because it is a monotonically increasing function and does not change the position of the maximum. We obtain

\begin{align*}
    \hat{x}^{\mathsf{MAP}}
    &= \argmax_{\tilde{x} \in \XX} 
        \Big[ \log p(y \mid \tilde{x}) + \log \Pr\{ \tilde{x}\} \Big]
     = \arg \! \max_{\tilde{x} \in \XX} 
        \Big[- \frac{|y - \tilde{x}|^2}{\sigmaW} 
         + \log \Pr\{ \tilde{x}\} \Big] \\
     &= \arg \! \min_{\tilde{x} \in \XX} 
        \Big[|y - \tilde{x}|^2 -\sigmaW \cdot \log \Pr\{ \tilde{x}\} \Big] \\
    \hat{x}^{\mathsf{ML}} 
    &= \arg \! \max_{\tilde{x} \in \mathbb{X}} \log p(y \mid \tilde{x})
     = \arg \! \min_{\tilde{x} \in \mathbb{X}} |y - \tilde{x}|^2 \; .
\end{align*}

For Gaussian noise, the ML detector searches for the data symbol with the smallest Euclidean distance to the received symbol $y$. The MAP detector extends the ML expression by the a priori probabilities of the hypotheses $\tilde{x}$. In the low SNR regime is the noise very strong (large variance $\sigma_{\cal W}^2$) and the prior knowledge dominates the detection. This is meaningful because the received value $y$ is very unreliable. In the high SNR regime, $\sigma_{\cal W}^2$ is small and the received symbol $y$ is trustful and dominates the decision.

An important parameter of the AWGN channel is the signal-to-noise ratio (SNR). Using the power spectral density $N_0$ of the noise and the energy $E_{\mathrm s}$ per symbol, it is defined to 

\begin{equation}
    \gamma = 
        \begin{cases}
            \frac{E_{\mathrm s}}{N_0} & \text{for complex-valued alphabets} \\ 
            \frac{E_{\mathrm s}}{N_0/2} & \text{for real-valued alphabets} 
    \end{cases}
\end{equation}

The average energy $E_{\mathrm{s}}$ per data symbol is related to the average energy $E_{\mathrm{b}}$ per information bit by $E_{\mathrm{s}}=m \cdot E_{\mathrm{b}}$ with $m$ being the number of bits per symbol. The corresponding signal-to-noise ratio becomes

\begin{equation}
    \frac{E_{\mathrm s}}{N_0} = m \cdot \frac{E_{\mathrm b}}{N_0}
    \quad \Rightarrow \quad
    10 \cdot \log_{10} \frac{E_{\mathrm s}}{N_0} 
    = 10 \cdot \log_{10} m + 10 \cdot \log_{10} \frac{E_{\mathrm b}}{N_0} .
\end{equation}

The energy efficiency $E_{\mathrm{b}}/N_0$ is particularly interesting in information theory.


## 2.2 Symbol Error Probabilities

A detection error occurs if the detector decides in favor to a wrong symbol, i.e. $\hat{x} \neq x$. This happens for Gaussian noise and ML detection when the competing symbol $\tilde{x} \neq x$ is closer to $y$ than the true symbol $x$. The probability for this event is called pairwise error probability and depends on the SNR and the modulation scheme. Its computation requires the integration of the likelihood function over a particular interval. Unfortunately, the Gaussian distribution cannot be integrated in closed form and numerical solutions have to be used. They are represented by the Gaussian error function $\mathrm{erf}(x)$ and its complement $\mathrm{erfc}(x)=1-\mathrm{erf}(x)$. They are defined by 

\begin{align*}
    \mathrm{erf}(x) &= \frac{2}{\sqrt{\pi}} \cdot \int_0^x e^{- \xi^2} \ d\xi \\
    \mathrm{erfc}(x) &= 1 - \mathrm{erf}(x)
    = \frac{2}{\sqrt{\pi}} \cdot \int_x^{\infty} e^{- \xi^2} \ d\xi .
\end{align*}

The symbol error probabilities for different linear modulation schemes depend on the SNR and the minimum squared Euclidean distance $\Delta_0^2$ of the signal alphabet. The latter is usually normalized on the average transmit power $\sigma_{\mathcal X}^2$.

\begin{equation}
    \Delta_0^2 = \min_{x,\tilde{x} \in \mathbb{X}, x \neq \tilde{x}} 
        \frac{|x - \tilde{x}|^2}{\sigma_{\cal X}^2}
\end{equation}

We obtain the following expressions for ASK, PSK and QAM.
\begin{align*}
    P_{\mathrm{s}}^{M-\mathsf{ASK}} 
    &= \frac{M-1}{M} \cdot \mathrm{erfc} \left( \sqrt{ \frac{3}{M^2-1} \frac{E_{\mathrm{s}}}{N_0} } \right)
     = \frac{M-1}{M} \cdot \mathrm{erfc} \left( \sqrt{ \left( \frac{\Delta_0}{2} \right)^2 \frac{E_{\mathrm{s}}}{N_0} } \right) \\
\end{align*}

## 2.3 Mutual Information as Function of SNR

In information theory, the mutual information is of major importance. For continuously distributed noise, differential entropies $h({\cal Y})$ have to be used. The mutual information is defined as the difference of two differential entropies

\begin{align}
    I({\cal X};{\cal Y}) &= h({\cal Y}) - h({\cal Y} \mid {\cal X}) \; .
\end{align}

The term $h({\cal Y} \mid {\cal X})$ denotes the irrelevance and equals the differential entropy of the noise $h({\cal W})$ and is constant for given noise power. For digital modulation schemes like ASK, QAM or PSK with discrete alphabets, the mutual information cannot be expressed by simple analytic expressions because the computation of $h({\cal Y})$ includes the logarithm of

\begin{align}
    p(y) = \sum_{x \in \mathbb{X}} p(y \mid x) \cdot \Pr \{ x\} .
\end{align}

The sum inside the logarithm cannot be simplified and the mutual information

\begin{align}
    I({\cal X};{\cal Y}) 
    = \sum_{x \in \mathbb{X}} \Pr\{ x \} \cdot 
      \int_{-\infty}^{\infty} p(y \mid x) \cdot \log_2 \frac{p(y \mid x)}{p(y)} \ dy
\end{align}

can only be determined by numerical integration. The mutual information's unit is bit/s/Hz. Hence, it describes a data rate normalized on the minimum required bandwidth. The mutual information is also called spectral efficiency.

Obviously, the mutual information depends on the channel SNR, i.e. it is a function of $E_{\mathrm s}/N_0$ and $E_{\mathrm b}/N_0$. The relationship between both in the information theoretic sense is 

\begin{equation}
    \frac{E_{\mathrm s}}{N_0} =  I({\cal X};{\cal Y}) \cdot \frac{E_{\mathrm b}}{N_0} \ .
\end{equation}
