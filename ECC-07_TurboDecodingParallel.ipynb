# 7. Turbo-Decoding of Parallel Concatenated Codes

In order to construct powerful codes with which the channel capacity defined by Shannon can be approached, code concatenation plays a crucial role. The basic concept has already been proposed in 1966 in the PhD thesis of David Forney. However, it took nearly three decades until a breakthrough in concatenated coding was achieved. With the invention of turbo codes in 1993, Shannon's capacity was approached by 0.5 dB only. In subsequent years, huge progress has been made in optimization and design of encoders and particularly the turbo decoding procedure. The latter has also been applied to different applications like interference cancellation, coded modulation, turbo equalization etc. EXIT and variance transfer charts have been developped and became powerful tools for the convergence analysis of the iterative decoding process. 

In this exercise, we will analyze simple parallel concatenated block codes and apply the turbo decoding principle. 


## 7.2 Parallel Code Concatenation

The parallel concatenation of two constituent codes $C_1$ and $C_2$ is depicted in the figure below. 

<img src="Figures/part4_parallel_concatenation_encoder.png" width=200 ref={ptp-link-cont}/>

The most important ingredient is the interleaver $\Pi$ between the two encoders. It permutes the input sequence $\uv$ such that $C_1$ and $C_2$ obtain the same information bits, just in different orders. The output sequences contain the corresponding parity bits $\cv_1$ and $\cv_2$, but not the information bits $\uv$. They are multiplexed with the information sequence to a final code sequence $\cv$. This multiplexing can include puncturing described by the matrix $\Pm$ in order to obtain a desired overall code rate $\Rc$. 

As the optimal ML decoding of the concatenated code is usually infeasible, a turbo decoding process is applied. Similar to the encoder, the turbo decoder consists of two decoders, one for each constituent encoder. They are connected via interleavers and de-interleavers as shown in the figure below. The decoders run alternatingly and exchange information such that the decoding result is improved with each iteration. This iterative decoding process is stopped if a suited criterion is fulfilled.

<img src="Figures/part4_parallel_turbo_decoder.png" width=600 ref={ptp-link-cont}/>

As decoding algorithms, so-called soft-in / soft-out decoding algorithms are required. They need LLRs at their inputs and deliver LLRs at their outputs. As can be seen from the figure above, $D_1$ and $D_2$ obtain LLRs $L(y_i | x_i)$ from the channel output. For systematic encoders, the decoder output consists of three additive parts.

\begin{align}
    L(x_i | \yv) 
    &= L(y_i | x_i) + \La(x_i) + \Le(x_i) \\
\text{intrinsic LLR } L(y_i | x_i) &= 4 \frac{\Es}{N_0} \cdot y_i \text{ for AWGN channels} \\
\text{a priori LLR } \La(x_i) &= \log \frac{\Pr\{ x_i=+1 \}}{\Pr\{ x_i=-1 \}} \\
\text{extrinsic LLR } \Le(x_i) &= \log \frac{\sum_{\xv \in \Gamma_i^{(+1)}} \prod_{j \neq i} p(y_j|x_j) \cdot \prod_{j \neq i} \Pr\{x_j\}}{\sum_{\xv \in \Gamma_i^{(-1)}} \prod_{j \neq i} p(y_j|x_j) \cdot \prod_{j \neq i} \Pr\{x_j\}}
\end{align}



Convergence of the iterative algorithm can be ensured if only the extrinsic LLRs $\Le(x_i)$ are exchanged between the decoders. Therefore, the subtraction points are inserted. They subtract the intrinsic and a priori LLRs from the decoder outputs such that only $\Le(x_i)$ remains. Interleaving and de-interleaving ensures that the extrinsic LLRs have the correct order for each decoder, i.e. the order matches to the one of the other inputs. With this concept, extrinsic LLRs of one decoder become a priori LLRs for the other decoder and hopefully improve the performance. 

Please be aware that for systematic encoding as assumed for the parallel code concatenation, decoder $D_2$ obtains intrinsic and a priori LLRs of the systematic information bits from $D_1$ while only the LLRs of its code bits $\cv_2$ are directly obtained from the channel output.
