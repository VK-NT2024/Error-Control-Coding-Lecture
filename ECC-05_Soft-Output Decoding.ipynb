# 5. Soft-Output Decoding

## 5.1 Log-Likelihood Ratios for Uncoded Binary Systems

In the context of concatenated codes, but as well for decoding Low Density Parity Check (LDPC) codes, the exchange of reliability information is essential. Reliability information means that not just the hard decision (sign of BPSK symbol), but also a measure about the reliability of this decision is provided. For binary codes, this so-called soft information can be represented by the log-likelihood ratio (LLR). For a single bit $u$ and its corresponding BPSK symbol $x = 1- 2u$, it is defined as

\begin{align}
    L_a(u) = \log \frac{\Pr\{ u=0 \}}{\Pr\{ u=1 \}} = \log \frac{\Pr\{ x=+1 \}}{\Pr\{ x=-1 \}} = L_a(x)
\end{align}

and only depends on the a priori distribution of $u$. Hence, $L_a(x)=L_a(u)$ is called a priori LLR. Regarding the transmission over an error-prone channel, the likelihood function $p(y | x)$ represents the channel statistics. The associated LLR becomes

\begin{align}
    L(y | x) &= \log \frac{ p(y | x=+1) }{ p(y | x=-1) } \ .
\end{align}

As $L(y | x)$ is based on the obervation of the received version of $x$, it is termed intrinsic information. As a third version of the LLRs, the a posteriori LLR has the form

\begin{align}
    L(x | y) 
    &= \log \frac{\Pr\{ x=+1 | y \}}{\Pr\{ x=-1 | y \}}
     = \log \frac{ p(y | x=+1) \Pr\{ x=+1 \} / p(y) }{ p(y | x=-1) \Pr\{ x=-1 \} / p(y)} 
     = \log \frac{ p(y | x=+1) }{ p(y | x=-1) \}} + \log \frac{ \Pr\{ x=+1 \} }{ \Pr\{ x=-1 \}} 
     = L(y | x) + L_a(x) \ .
\end{align}

Hence, the a posteriori LLR is the sum of the intrinsic LLR and the a priori LLR. LLRs contain two types of information:

* the sign represents the hard decision
* the magnitude represents the reliability of the hard decision.

The probability of a correct decision can be expressed by $\Pr \{ \hat{x} \text{ is correct} \} = 1 - \Ps = \frac{1}{1+e^{-|L(x|y|}}$. Consequently, the probability of a decision error becomes $\Pr \{ \hat{x} \text{ is wrong} \} = \Ps = \frac{1}{1+e^{|L(x|y|}}$ (see also Exercise 4.4). For the intrinsic LLR, the following list presents expressions for common communication channels.

* Binary Erasure Channel (BEC) with erasure probability $\Pe$
\begin{equation}
   L(y|x) = \begin{cases} \log \frac{1-\Pe}{0} = + \infty & \text{ for } y=+1 \\ \log \frac{\Pe}{\Pe} = 0 & \text{ for } y=0 \\ \log \frac{1-\Pe}{0} = + \infty \log \frac{0}{1-\Pe} = - \infty & \text{ for } y=-1 \end{cases}
\end{equation}
* Binary Symmetric Channel (BSC) with error probability $\Pe$
\begin{equation}

## Exercise 5.2 Log-Likelihood Ratios for Repetition Codes

For repetition codes, the intrinsic LLRs $L(y_i|x_i)$ must be summed for soft-output decoding. The a posteriori LLR becomes 
\begin{align}
    L(u|\yv) 
    &= \log \frac{\Pr\{u=0|\yv\}}{\Pr\{u=1|\yv\}} = \log \frac{p(\yv|u=0) \cdot \Pr\{u=0\}}{p(\yv|u=1) \cdot \Pr\{u=1\}} \\
    &= \log \frac{\Pr\{u=0\}}{\Pr\{u=1\}} + \log \prod_{i=0}^{n-1} \frac{p(y_i|x_i=+1)}{p(y_i|x_i=-1)} \\
    &= \La(u) + \sum_{i=0}^{n-1} L(y_i | x_i)
\end{align}

## Exercise 5.3: Soft-Output Decoding of Single Parity Check Code

For single parity check codes, the intrinsic LLRs $L(y_j|x)$ cannot be summed as for the repetition code. Instead, the $L$-algebra must be applied. At the encoder, the single parity bit $p$ is computed by $p=u_0 \oplus \cdots \oplus u_{k-1} = \sum_{i=0}^{k-1} u_i \mod 2$. Resolving this equation w.r.t. an arbitrary information bit $u_j$, we obtain $u_j = \sum_{i=0, i \neq j}^{k-1} u_i \mod 2 \oplus p$. With the BPSK mappings $u_i \to x_i$ and $p \to x_{n-1}$, the a posteriori LLR for bit $u_j$ (symbol $x_j$) becomes

\begin{align}
    L(x_j|\yv) 
    &= \log \frac{\Pr\{x_j=+1|\yv\}}{\Pr\{x_j=-1|\yv\}}
     = \log \frac{\sum_{\xv \in \Gamma_j^{(+1)}} \Pr\{\xv|\yv\}}{\sum_{\xv \in \Gamma_j^{(-1)}} \Pr\{\xv|\yv\}} 
     = \log \frac{\sum_{\xv \in \Gamma_j^{(+1)}} p(\yv|\xv) \cdot \Pr\{\xv\}}{\sum_{\xv \in \Gamma_j^{(-1)}} p(\yv|\xv) \cdot \Pr\{\xv\}} 
\end{align}

The subsets $\Gamma_j^{(\pm 1)}$ represent the set of codewords having the symbol $x_j = \pm 1$ ($u_j = 0,1$) at position $j$. For memoryless channels like the AWGN channel, the likelihood functions can be factorized to $p(\yv|\xv) = \prod_{i=0}^{n-1} p(y_i|x_i)$. Moreover, the information bits are usually assumed to be independent leading to $\Pr\{\xv\} = \Pr\{\uv(\xv)\} = \prod_{i=0}^{k-1} \Pr\{u_i\}$. With these simplifications, the above expression becomes

\begin{align}
    L(x_j|\yv) 
     &= \log \frac{\sum_{\xv \in \Gamma_j^{(+1)}} \prod_{i=0}^{n-1} p(y_i|x_i) \cdot \prod_{i=0}^{k-1} \Pr\{x_i\}}
                  {\sum_{\xv \in \Gamma_j^{(-1)}} \prod_{i=0}^{n-1} p(y_i|x_i) \cdot \prod_{i=0}^{k-1} \Pr\{x_i\}} 
      = \log \frac{p(y_j|x_j=+1) \cdot \Pr\{x_j=+1\} \cdot \sum_{\xv \in \Gamma_j^{(+1)}} \prod_{i \neq j} p(y_i|x_i) \cdot \prod_{i \neq j} \Pr\{x_i\}}
                  {p(y_j|x_j=-1) \cdot \Pr\{x_j=-1\} \cdot \sum_{\xv \in \Gamma_j^{(-1)}} \prod_{i \neq j} p(y_i|x_i) \cdot \prod_{i \neq j} \Pr\{x_i\}} \\
     &= L(y_j|x_j) + \La(x_j) + \log \frac{\sum_{\xv \in \Gamma_j^{(+1)}} \prod_{i \neq j} p(y_i|x_i) \cdot \prod_{i \neq j} \Pr\{x_i\}}
                                          {\sum_{\xv \in \Gamma_j^{(-1)}} \prod_{i \neq j} p(y_i|x_i) \cdot \prod_{i \neq j} \Pr\{x_i\}} \\
     &= L(y_j|x_j) + \La(x_j) + \Le(x_j) \ .
\end{align}

For the single parity check code and uniformly distributed information bits with $\Pr\{x_i=+1\} = \Pr\{x_i=-1\}=\frac{1}{2}$, the computation of the extrinsic LLRs $\Le(x_j)$ simplifies to

\begin{align}
    \Le(x_j) 
     &= \log \frac{1 + \prod_{j \neq i} \tanh \left( \frac{L(x_i)}{2} \right)}{1 - \prod_{j \neq i} \tanh \left( \frac{L(x_i)}{2} \right)} 
      = 2 \artanh \left[ \prod_{i \neq j} \tanh \left( \frac{L(x_i)}{2} \right) \right] \ .
\end{align}
