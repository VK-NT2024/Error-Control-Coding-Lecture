# 8. Extrinsic Information Transfer (EXIT) Charts for Parallel Concatenated Blockcodes

## 8.1 Basic idea<a class="anchor" id="BasicIdea"></a>

Whenever iterative algorithms are applied, the question of convergence arises. Particularly, it has to be analyzed 

* under which conditions an iterative algorithm will converge
* to which solution the algorithm will converge. 

This analysis is also required for turbo decoding. As coding theory is tightly coupled with information theory, it looks quite natural to analyze the exchange of mutual information between constituent decoders. This idea was first proposed by Stephan ten Brink<a name="tenBrink-1"></a>[<sup>[1]</sup>](#tenBrink-1) <a name="tenBrink-2"></a>[<sup>[2]</sup>](#tenBrink-2) <a name="tenBrink-3"></a>[<sup>[3]</sup>](#tenBrink-3) <a name="tenBrink-4"></a>[<sup>[4]</sup>](#tenBrink-4). As we have learned in the previous exercises, only extrinsic LLRs are exchanged between decoders. Consequently, the *extrinsic mutual information* becomes the crucial performance measure in turbo decoding processes. Unfortunately, the LLRs and also the corresponding mutual information cannot be determined analytically. Instead, they have to be numerically computed. Doing this inside of turbo decoding iterations is computational expensive because the whole decoding process has to be repeated for all possible combinations of codes and decoding algorithms. However, extensive Monte-Carlo simulations revealed that the extrinsic LLRs at a decoder output are approximately Gaussian distribued. Based on this observation, a semi-analytical approach has been developed, the EXtrinsic Information Transfer (EXIT) chart method. Essentially, it analyzes each decoder separately and subsequently combine the results to an EXIT chart. This saves computational costs because the transfer charts of different codes can be combined without simulating the whole turbo decoding process. The basic procedure is summarized below for parallel concatenated codes.

EXIT charts have been successfully applied in many different areas. Besides the concatenation of decoders, bit interleaved coded modulation (BICM) can be analyzed by this approach as well. Coding and equalization or multi-user detection are further examples. The EXIT chart analysis tightly predicts the behavior of turbo decoders if the block length are sufficiently large. If block lengths become too short, the statistical deviations increase and a tight prediction is no longer possible. This will also be observed in this exercise. Equivalent strategies like variance transfer charts exist and provide equivalent results.


**Parallel Code Concatenation**

For parallel concatenated codes, the turbo decoder is depicted in the figure below. The relevant mutual information terms are included. Besides the intrinsic LLRs $L(y|x) = L_\mathrm{ch} y$ coming from the channel output, each decoder also has a second input with a priori LLRs $L_{\mathrm{a},i}(u)$ coming from the extrinsic output $L_{\mathrm{e},j \neq i}(u)$ of the other decoder. As a performance measure, the mutual information between these LLRs and the true information bits $u$ has to be determined. $I_{\mathrm{e},i}=I(u;L_{\mathrm{e},i}(u))$ denotes the average extrinsic mutual information between $u$ and the extrinsic LLRs of decoder $i$. Correspondingly, $I_{\mathrm{a},i}=I(u;L_{\mathrm{a},i}(u))$ represents the average extrinsic mutual information between $u$ and the a priori LLRs of decoder $i$.

**Transfer Chart of Single Decoder**

As mentioned above, we now concentrate the analysis on a single decoder according to the figure below. For a particular code, a Monte-Carlo simulation is implemented assuming BPSK transmission over an AWGN channel. The received signal $y = x + w$ is corrupted by AWGN with a specified SNR. The corresponding mutual information $I(\Xrv;\Yrv)$ represents the maximum mutual information the coded system can achieve. Appropriate scaling the the channel reliability $L_\mathrm{ch}$ delivers the intrinsic LLRs $L(y|x)$. The decoder's a priori LLRs are generated artificially by mapping the information bits to BPSK symbols and adding AWGN with a specific variance. This yields

\begin{equation}
    \La(u) = a \cdot (1-2 \cdot u) + w_\mathrm{a} \qquad \text{with} \qquad w_\mathrm{a} \sim {\Nrv} (0,\sigma_\mathrm{a}^2) \ . 
\end{equation}

The factor $a$ in front of the data symbol cannot be chosen arbitrary but is a function of the artificial noise variance $\sigma_{\mathrm{a}}^2$ (see Exercise 8.1). The a priori mutual information between information bits and $\La(u)$ the decoder's a priori input has to be computed by solving

\begin{equation}
    \Ia = 1 + \frac{1}{2} \cdot \sum_{\mu=0}^{1} \int_{-\infty}^{\infty} p(\La|u=\mu) \log_2 \bigg[ \frac{p(\La|u=\mu)}{p(\La|u=0) + p(\La|u=1)} \bigg] d\La 
\end{equation}

numerically. There is a mononotonic relationship between $I_\mathrm{a}$ and $\sigma_\mathrm{a}^2$ such that $\sigma_\mathrm{a}^2$ can be easily altered in a systematic manner in order to obtain dedicated values for $I_\mathrm{a}$. Now, the decoder is analyzed for different channel SNRs and different qualities of the a priori LLRs measured by $I_\mathrm{a}$. For each combination of SNR and $I_\mathrm{a}$, the extrinsic LLRs $L_\mathrm{e}(u)$ and the associated extrinsic mutual information $I_\mathrm{e}$ are determined. This can be accomplished in two different ways.  

* Computation by using histograms (estimates of probability density function)
\begin{align*}
    \hat{I}_{\mathrm{e}} = 1 + \frac{1}{2} 
        &\cdot \sum_{\mu=0}^{1} \int \limits_{-\infty}^{\infty} \hat{p}(L_{\mathrm{e}} \mid u=\mu)
         \cdot \log_2 \left[ \frac{\hat{p}(L_{\mathrm{e}} \mid u=\mu)}{\hat{p}(L_{\mathrm{e}} \mid u=0)+ \hat{p}(L_{\mathrm{e}} \mid u=1)} \right] dL_{\mathrm{e}} 
\end{align*}
* Computation using reliabilities of bits (magnitudes of LLRs)
\begin{align}
    \hat{I}_{\mathrm{e}} 
        = \frac{1}{L} \sum_{i=1}^{L} \frac{1}{1+e^{|L_{\mathrm{e},i}|}} \cdot 
                  \log_2 \frac{2}{1+e^{|L_{\mathrm{e},i}|}} + \frac{1}{1+e^{-|L_{\mathrm{e},i}|}} \cdot \log_2 \frac{2}{1+e^{-|L_{\mathrm{e},i}|}} 
\end{align}

<img src="Figures/part4_exit_single_decoder.png" width=600 ref={ptp-link-cont}/>

As the histogram based method requires to choose the number of bins and the bin width appropriately, it is beneficial to use the LLR-based approach. However, this only delivers accurate estimates for $I_\mathrm{e}$ if true LLRs are provided by the decoder. For approximate algorithms like the Max-Log-MAP, this approach can become inaccurate. For the outer decoder of a serial concatenation, the procedure is the same. The only difference is the consideration of the code bits instead of the information bits. Furthermore, as the decoder output is independent of the channel SNR, only the a priori mutual information has to be altered.


**Concergence Analysis of Turbo Decoder**

When the analysis of the constituent decoders as described above has been finished, the transfer charts can be drawn into a single diagram. However, it has to be considered that the extrinisic mutual information of one decoder becomes the a priori mutual information of the other decoder. Due to $I_{\mathrm{e},2}=I_{\mathrm{a},1}$ and $I_{\mathrm{e},1}=I_{\mathrm{a},2}$, the axes for one of the decoders have to be exchanged.
