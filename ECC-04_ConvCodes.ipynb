# 4. Convolutional Codes

## 4.1 Definitions

**Encoder:**
An $(n,k,\Lc)$ convolutional encoder can be implemented by a linear shift register having $n$ output symbols per time instance and $k$ input symbols. An example of a binary encoder with $k=1$ and $n=2$ is shown in [the figure below](ptp-link-conv_encoder). The memory of this famous (7,5,3) encoder is $m=2$ and its constraint length is $\Lc=m+1=3$. Similar to the generator matrix of linear block codes, the convolutional encoder is completely described by its generators $\gv_1=[ 1 \ 1 \ 1]$ and $\gv_2=[ 1 \ 0 \ 1]$. Obviously, the two output streams $c_{\ell,1}$ and $c_{\ell,2}$ are obtained by convolving the input sequence $u_\ell$ with these generators. 

\begin{align}
    c_{\ell,1} &= u_\ell \ast g_{\ell,1} \mod 2 \\
    c_{\ell,2} &= u_\ell \ast g_{\ell,2} \mod 2
\end{align}

The two output sequences $c_{\ell,1}$ and $c_{\ell,2}$ are usually demultiplexed to a single sequence according to $\ldots \ , c_{\ell-1,1}, c_{\ell-1,2}, \ c_{\ell,1}, c_{\ell,2}, \ c_{\ell+1,1}, c_{\ell+1,2}, \ \ldots$. We will denote the entire code sequence by a vector $\cv$. It contains $L$ code words $\cv_\ell \in \mathrm{GF}(2)^n$ each consisting of $n$ bits. 

<img src="Figures/part3_conv_encoder.png" width=400 ref={ptp-link-conv_encoder}/>

The example above represents a nonsystematic encoder because the information sequence $u_\ell$ cannot be explicitly found in the output sequence.

**Duality between block and convolutional codes:**
Due to the convolutional encoder's memory, it can be interpreted as a linear block code with memory. Contrarily, block codes can be interpreted as memoryless convolutional codes. If the entire output sequence is interpreted as a single code word $\cv$, the corresponding generator matrix has a Toeplitz structure (due to the convolution).


**Code rate:**
The code rate of a convolutional encoder is defined as

\begin{equation}
  \Rc = \frac{k}{n} 
\end{equation}

and becomes $\Rc=\frac{1}{2}$ for the example above.

**Distance Properties:**
An important performance measure of convolutional codes is the free distance $\df$. It represents the smallest Hamming distance between any two code sequences $\cv^{(i)}$ and $\cv^{(j)}$. 

\begin{equation}
    \df = \min_{\cv^{(i)}, \cv^{(j)} \in \Gamma, i \neq j} \dH(\cv{(i)}, \cv^{(j)})
\end{equation}


**Decoding and Error Correction:**
Due to the encoder's memory, successive codewords $\cv_i$ depend on each other. Consequently, a decoder cannot decode a single codeword $\cv_i$ but has to decode a complete sequence of codewords $\cv = [ \cv_0, \cv_1, \ldots , \cv_L]$. For the transmission over an AWGN channel, a maximum likelihood decoder determines the hypothesis that maximizes the likelihood function.

\begin{align}
    \hat{\cv} 
    &= \argmax_{\tilde{\cv}} p(\yv | \cv)
     = \argmax_{\tilde{\cv}} \log p(\yv | \cv) 
    \condequal_{\mathrm{AWGN}} \argmin_{\tilde{\cv}} \| \yv - \xv \|^2
\end{align}

The sequence represented by the vector $\xv = \mathbf{1} - 2 \cv$ denotes the BPSK modulated version of the bit sequence $\cv$. An efficient algorithm that solves the above optimization problem is the Viterbi algorithm. It is based on the trellis representation of the code and has a computational complexity that scales linearly with the length of $\uv$ but still exponentially with the encoders memory $m$. The number of errors the Viterbi decoder can correct cannot be easily determined. Certainly, $t= \left\lfloor \frac{\df-1}{2} \right\rfloor$ errors can always be corrected. However, if the bit errors are not bursty but spread over the sequence such that they can be seen as independent bit errors, the decoder might correct much more than $t$ errors. 

**Trellis Truncation and Termination:**
The convolutional encoder can be seen as a finite state machine having $2^m$ states with $m$ being the encoder's memory. As the corresponding state diagram does not contain any temporal information, the trellis diagram is an appropriate choice for the graphical code representation. Here, the states are vertically arranged, the horizontal direction represents the time axis. Terminated and truncated trellises are distinguished. Terminated trellises start and end in the same state, which is usually the all-zero-state. This is accomplished by appending tail bits to the information sequence $\uv$. For an encoder with memory $m$, $m$ tail bits are needed to force the encoder to return into the all-zero-state. Since the tail bits do not represent information, they reduce the code rate. Assuming $K$ information bits, we obtain $L$ codeword each of length $n$ and the coderate for terminated trellis becomes

\begin{equation}
    \Rc^{\mathrm(term)} = \frac{K}{(K+m) \cdot n} = \Rc \cdot \frac{K}{K+m} \ . 
\end{equation}

We see that the original code rate $\Rc$ is reduced by a factor $\frac{K}{K+m}$. For $K \gg m$, the rate reduction can be neglected. Contrarily, it becomes significant for small $k$ like for control channels in mobile radio systems. Here, truncated trellises, which stop with the last information bit, are preferable. Consequently, the final state in the trellis is unknown to the decoder. This leads to unreliable estimates of the last bits in a sequence.
