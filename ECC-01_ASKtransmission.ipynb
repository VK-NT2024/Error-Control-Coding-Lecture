# 1. Simulation of Uncoded ASK Transmission

## 1.1 System Model<a class="anchor" id="SystemModel"></a>

This Jupyter notebook considers an uncoded ASK transmission over an AWGN channel. The simplified continuous-time model of the point-to-point link is illustrated in [the figure below](ptp-link-cont).

<img src="Figures/part1_uncoded_ASK_transmission_continuous.png" width=800 ref={ptp-link-cont}/>

The ASK (Amplitude Shift Keying) mapper maps $m$ information bits $u[k]$ onto an ASK symbols $x[i] \in \mathbb{X}$. The alphabet $\mathbb{X}$ contains all $M=2^m$ ASK symbols. The symbol $x[i]$ are passed to the transmit filter $g_{\mathrm{T}_\mathrm{x}}(t)$ to obtain a physical signal 

\begin{equation}
    x(t) = T_\mathrm{s} \cdot \sum_i x[i] \cdot g_{\mathrm{T}_{\mathrm{x}}}(t-i T_\mathrm{s})
\end{equation}

being continuous in time and amplitude. The transmission is corrupted by additive white Gaussian noise $w(t)$ delivering $y(t) = x(t) + w(t)$ at the receiver. Here, the receive filter $g_{\mathrm{R}_\mathrm{x}}(t)$ performs lowpass filtering followed by sampling at symbol rate $T_\mathrm{s}^{-1}$. 

\begin{align}
    y[i] &= \left. \left( g_{\mathrm{R}_\mathrm{x}}(t) \ast y(t) \right) \right|_{t = i T_\mathrm{s}}
          = T_\mathrm{s} \cdot \sum_l x[l] \cdot g_{\mathrm{R}_{\mathrm{x}}}(t) \ast g_{\mathrm{T}_{\mathrm{x}}}(t-l T_\mathrm{s}) \big|_{t = i T_\mathrm{s}}  + g_{\mathrm{R}_{\mathrm{x}}}(t) \ast w(t) \big|_{t = i T_\mathrm{s}}\\
        &= T_\mathrm{s} \cdot \sum_l x[l] \cdot g(t-l T_\mathrm{s}) \big|_{t = i T_\mathrm{s}} + w[i]
\end{align}

The term $w[i]$ represents the filtered noise sampled at $t=i T_\mathrm{s}$. It remains white and Gaussian distributed if the receive filter fulfills certain conditions, e.g. its squared transfer function has a symmetric Nyquist slope. In order to maximize the signal to noise ratio (SNR), the receive filter is matched to the transmit filter, i.e. $g_{\mathrm{R}_\mathrm{x}}(t)=g_{\mathrm{T}_\mathrm{x}}(-t)$ holds. Transmit and receive filter are usually designed sucht that the overall impulse response $g(t) = g_{\mathrm{R}_\mathrm{x}}(t) \ast g_{\mathrm{T}_\mathrm{x}}(t)$ fulfills the first Nyquist criterion, i.e.

\begin{equation}
    g(i T_\mathrm{s}) = 
    \begin{cases}
        T_\mathrm{s}^{-1} & \text{for } i=0 \\
        0 & \text{else} \; .
    \end{cases}
\end{equation}

If sampling is performed at ideal times $i T_\mathrm{s}$, successive symbols do not interfere with each other (no intersymbol interference, ISI) and one received symbol $y[i]$ only depends on the associated data symbol $x[i]$. 

\begin{equation}
    y[i] = T_\mathrm{s} \cdot \sum_l x[l] \cdot g(l T_\mathrm{s} - i T_\mathrm{s}) + w[i] = x[i] + w[i] 
\end{equation}

and the discrete-time system model shown in [the figure below](ptp-link-discrete).

## 1.2 Maximum-Likelihood (ML) Detection 

The optimal Maximum-Likelihood detector works on a symbol-by-symbol basis because no ISI affects the transmission and the noise is white. Generally, it maximizes the likelihood function and yields the estimate

\begin{equation}
    \hat{x}^{\mathsf{ML}}[i] 
        = \arg \! \max_{\tilde{x} \in \mathbb{X}} p(y[i] \mid \tilde{x})
        = \arg \! \max_{\tilde{x} \in \mathbb{X}} \log p(y[i] \mid \tilde{x}) \; .
\end{equation}

For Gaussian noise, it determines that symbol $\hat{x}[i] \in \mathbb{X}$ which has the smallest Euclidean distance to the received symbol $y[i]$. In tis case, we obtain

\begin{equation}
    \hat{x}^{\mathsf{ML}}[i] = \arg \! \min_{\tilde{x} \in \mathbb{X}} |y[i] - \tilde{x}|^2 \; .
\end{equation}

The performance of digital communication systems is usually measured by error probabilities. For the uncoded transmission, we distinguish bit and symbol error probabilities. In a Monte Carlo simulation, they can be determined by comparing the transmitted bits/symbols with the detected ones. Counting the number of different bits/symbols and dividing by the total number of transmitted bits/symbols provides the corresponding error rates. With the indicator function 

\begin{equation}
    \delta(x) =  
        \begin{cases} 
            1 & \text{for } x \neq 0 \\ 
            0 & \text{for } x=0 
        \end{cases}
\end{equation}

the error rates can be expressed by

\begin{align}
    P_{\mathrm{s}} &= \frac{ 1 }{ N_{\mathrm{x}} } \cdot \sum_{i=0}^{N_{\mathrm{x}}-1} \delta(x[i] - \hat{x}[i]) \\
    P_{\mathrm{b}} &= \frac{ 1 }{ N_{\mathrm{u}} } \cdot \sum_{k=0}^{N_{\mathrm{u}}-1} \delta(u[k] - \hat{u}[k]) \approx \frac{P_{\mathrm{s}}}{m} \ .
\end{align}

$N_{\mathrm{x}}$ and $N_{\mathrm{u}}$ denote the number of data symbols and information bits, respectively. Certainly, $P_{\mathrm{s}}$ and $P_{\mathrm{b}}$ are identical for binary modulation schemes.
